```python

#텍스트 파일 불러오기
sample = open('/content/drive/MyDrive/Colab Notebooks/file/wordcloud_data.txt').read()

sample_nouns = okt.nouns(sample)
sample_nouns = " ".join(sample_nouns)

#단어구름 출력
wc = WordCloud(width=1200,
               height=1200,
               font_path='/content/drive/MyDrive/Colab Notebooks/file/malgun.ttf',
               background_color = 'white').generate(sample_nouns)
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
